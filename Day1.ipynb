2.1.4 Customize the index

```python
data1 = pd.Series([1,2,3,4], index=['a','b','c','d'])

data2 = pd.Series({'a':1,'b':2,'c':3,'d':4})

print(data1)
print(data2)
```

```python
data1 = pd.Series([1,2,3,4], index=['a','b','c','d'])
data2 = pd.Series({'a':1, 'b':2, 'c':3, 'd':4})
print(data1['a'])
print(data2['b'])
```

2.2 List and List operations by Pandas

```python
x = pd.Index([2,3,5,7,9,11])
print(x)
```

```python
import pandas as pd
a = pd.Index([1,3,5,7,9])
b= pd.Index([2,3,5,7,11])
print(a)
print(b)
print(a & b)
```

```python
print(a | b)
```

```python
print(a ^ b)
```
